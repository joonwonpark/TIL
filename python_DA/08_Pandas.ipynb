```python
##Database에서 sql로 데이터를 추출해서 pandas의 DataFrame으로 생성 
#사용하는 DBMS는 MySQL, 사용하는 데이터는 book table
#데이터베이스는 강사용 컴퓨터를 이용 :"70.12.116.160"
import pymysql.cursors   #MySQL DBMS를 사용하기 위한 module로딩
import numpy as np
import pandas as pd

#데이터베이스 connection
#conn = pymysql.connect(host="70.12.116.160")  강사용 컴터에 연결할거면
conn = pymysql.connect(host="localhost",
                      user="data",
                      password="data",
                      db="library",
                      charset="utf8")
keyword="여행"
#database에서 data를 가져오기 위한 SQL을 작성
#책 제목에 keyword가 들어있는 행을 찾아서 제목, 가격, 저자정보를 가져와요
sql=("select btitle,bprice,bauthor " +\
     "from book "+\
     "where btitle like '%{}%'".format(keyword))
try:
    df=pd.read_sql(sql, conn)
    display(df)
    #정상적으로 DataFrame을 얻어왔으면
    #JSon File로 저장 
    df.to_json("./data/book_columns.json",orient="columns")
    df.to_json("./data/book_records.json",orient="records")
    df.to_json("./data/book_index.json",orient="index")
    df.to_json("./data/book_values.json",orient="values")
except Exception as inst:
    print(inst)
finally:
    conn.close()   #처리가 성공하든 그렇지 않든 상관없이 무조건 database를 닫아요 
```

```python
#영화진흥위원회의 일일 박스오피스 정보를 이용해서 DataFrame을 만들어 보아요 
#접속해서 JSON을 얻어내면 그걸 분석해서 내가 원하는 형태의 DataFrame을 생성 
import pprint
import numpy as np
import pandas as pd
import urllib  #network 접속하기 위해서 필요 
import json  #openAPi의 결과가 JSON
            #json 읽어 들이기 위한 module
    
movie_url="http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/"+\
            "searchDailyBoxOfficeList.json?"+\
            "key={}&targetDt={}".format("8da0ed317a2bcbf29e42f7a499f96410",
                                       "20191120")

page=urllib.request.urlopen(movie_url)    #결과 JSON문자열이 들어있는 page객체
json_page=json.loads(page.read())
print(type(json_page))   ##Json 문자열을 dict로 변환 

movie_data={}
rank_list=list()
title_list=list()
sales_list=list()
for m_dict in json_page["boxOfficeResult"]["dailyBoxOfficeList"]:
    rank_list.append(m_dict["rank"])
    title_list.append(m_dict["movieNm"])
    sales_list.append("{:,}".format(int(m_dict["salesAmt"])))
movie_data["순위"]=rank_list
movie_data["영화제목"]=title_list
movie_data["당일매출액"]=sales_list
df=pd.DataFrame(movie_data)
display(df)
```

```python
#평균, 편차, 분산, 표준편차, 공분산, 상관계수
#평균(mean)
#편차(deviation): 확률변수 x와 mean의 차이   편차의 합은 0
#분산(variance):데이터의 흩어짐 정도를 알기위해서 사용하는 값으로 편차의 제곱의 평균
# 표준편차(standard deviation):분산의 제곱근   
#     분산의 단위 문제를 해결하고자 사용 
import numpy as np
arr=np.array([4,6,1,3,8,8],dtype=np.int32)
print(arr)
print(arr.sum())   #합계
print(np.sum(arr))
print(arr.mean())
print(arr.var())  #분산
print(arr.std())#표준편차

```

```python
#공분산
# 공분산: 두개의 확률변수X,Y가 있을때
#     이 두개의 확률변수의 관계를보여줄 때 사용해요
# 공분산은 두 확률변수의 편차의 곱의 평균으로 표현
#데이터가 평균으로부터 얼마나 떨어져있느냐를 표현 
#그래프를 그려보아요
import numpy as np
import matplotlib.pyplot as plt
#만약 module이 없다고 error가 뜨면 conda install matplotlib
np.random.seed(1)  #항상 같은 난수가 발생하도록 seed 처리 
x=np.random.randint(-20,20,(10,))
y=np.random.randint(-10,10,(10,))
x_mean=x.mean()
y_mean=y.mean()

x_deviation = x - x_mean
y_deviation = y - y_mean
print(x_deviation)
print(y_deviation)

result = 0
for i in range(10):
    result += (x_deviation[i]*y_deviation[i])
       
my_cov = result / 9 # 모 공분산 : 원래 공식
                     # 표준 공분산 : (n-1) 로 나누어요 추정치가 좋아요
print(my_cov)
print(np.cov(x,y))

#scatter(산점도) 표현해보아요
plt.scatter(x,y,color="red")
plt.scatter(x_mean,y_mean,color="blue")
plt.show()
```

```python
# DataFrame에서 sum함수와 mean함수의 사용
import numpy as np
import pandas as pd

# 중첩 리스트를 이용해서 DataFrame을 생성해보아요
data = [[2,np.nan],
       [7,-3],
       [np.nan,np.nan],
        [1,-2]]
df = pd.DataFrame(data,
                 columns=["one","two"],
                 index=["a","b","c","d"])
# display(df)
# df.sum() # DataFrmae에서는 sum()함수에 대해 axis를 주지 않으면
#          # axis =0 으로 간주 (열의합), NaN을 0으로 간주
# df.sum(axis=1)
# df["two"].sum() #-5.0
# df.loc["b"].sum() # 4.0

# df.mean(axis=0,skipna=False) # NaN은 실수처리, 모든 연산결과는 NaN
# df.mean(axis=0,skipna=True) # NaN있는항은 없는거처럼 간주

# "one" column의 결측값은 "one" column의 평균으로 
# "two" column의 결측값은 "two" column의 최솟값으로 대체
one_avg = df["one"].mean(axis=0,skipna=True)
two_avg = df["two"].min(axis=0)

df["one"] = df["one"].fillna(value = one_avg)
df["two"] = df["two"].fillna(value = two_avg)
display(df)
```

```python
# DataFrame Merge
# inner join, outer join(full) left join, right join
import numpy as np
import pandas as pd

data1 = {
    "학번" : [1,2,3,4],
    "이름" : ["홍길동", "최길동", "아이유","김연아"],
    "학년" : [2,3,3,1]
}

data2 = {
    "학번" : [1,2,4,5],
    "학과" : ["컴퓨터","미술","철학","사회"],
    "학점" : [1.3,3.5,4.3,2.3]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)

pd.merge(df1,df2,on="학번",how="inner")
pd.merge(df1,df2,on="학번",how="outer")
pd.merge(df1,df2,on="학번",how="left")
pd.merge(df1,df2,on="학번",how="right")
```

```python
import operator
import pandas as pd
df=pd.read_csv("./data/mpg.csv")
df_mpg=pd.DataFrame(df)
#isplay(df_mpg)
# 1. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비)가 평균적으로 더 높은지 확인하세요.
d4=df.loc[df["displ"]<=4,["hwy"]]
d4.mean()  # 25.96319
d5=df.loc[df["displ"]>=5,["hwy"]]
d5.mean()       # 18.078947
# 2. 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고  한다. "audi"와 "toyota" 중 어느 manufacturer(제조회사)의 
#       cty(도시 연비)가 평균적으로 더 높은지 확인하세요.       
dfa=df.loc[df["manufacturer"]=="audi",["cty"]]     
dfa.mean()   #17.611111
dft=df.loc[df["manufacturer"]=="toyota",["cty"]]    
dft.mean()  # 18.529412
#  3. "chevrolet", "ford", "honda" 자동차의 고속도로 연비 평균을 알아보려고 한다. 이 회사들의 데이터를 추출한 후 
#  hwy(고속도로 연비) 전체 평균을 구하세요
df3=df.loc[(df["manufacturer"]=="chevrolet")|(df["manufacturer"]=="ford")|(df["manufacturer"]=="honda"),"hwy"]
df3.mean()   #22.50943396226415
# 4. "audi"에서 생산한 자동차 중에 어떤 자동차 모델의 hwy(고속도로 연비)가 높은지 알아보려고 한다. 
#       "audi"에서 생산한 자동차 중 hwy가 1~5위에 해당하는 자동차의 데이터를 출력하세요.
df4=df.loc[df["manufacturer"]=="audi"]
df4.sort_values(axis=0,by=["hwy"],ascending=False).head(5)
# 5. mpg 데이터는 연비를 나타내는 변수가 2개입니다. 두 변수를 각각 활용하는 대신 하나의 통합 연비 변수를 만들어 사용하려 합니다. 평균 연비 변수는 
# 두 연비(고속도로와 도시)의 평균을 이용합니다. 회사별로 "suv" 자동차의 평균 연비를 구한후 내림차순으로 정렬한 후 1~5위까지 데이터를 출력하세요.
df5=df.loc[df["class"]=="suv",:]
df5['avg']=(df5['hwy']+df5['cty'])/2
df5=df5.groupby(['manufacturer'], as_index=False).mean()
df5.sort_values(by=["avg"],ascending=False).head(5)
# 6. mpg 데이터의 class는 "suv", "compact" 등 자동차의 특징에 따라 일곱 종류로 분류한 변수입니다. 어떤 차종의 도시 연비가 높은지 비교하려 합니다. 
# class별 cty 평균을 구하고 cty 평균이 높은 순으로 정렬해 출력하세요.
df6=df
df6=df6.groupby(['class'], as_index=False).mean()
df6.sort_values(by=["cty"],ascending=False)
# 7. 어떤 회사 자동차의 hwy(고속도로 연비)가 가장 높은지  알아보려 합니다. hwy(고속도로 연비) 평균이 가장 높은 회사 세 곳을 출력하세요.
df7=df
df7=df7.groupby(['manufacturer'], as_index=False).mean()
df7.sort_values(by=["hwy"],ascending=False).head(3)
# 8. 어떤 회사에서 "compact" 차종을 가장 많이 생산하는지 알아보려고 합니다. 각 회사별 "compact" 차종 수를 내림차순으로 정렬해 출력하세요.
df8=df.loc[df["class"]=="compact",:]
df8["class"].groupby(df8['manufacturer']).count().sort_values(ascending=False)
```
